## Convert Strings to Numbers

```python
df = pd.DataFrame({
    'col_1': ['1.1', '2.2', '3.3'],
    'col_2': ['4', '5', '6'],
    'col_3': ['7', '8', '-']
})
df
```

```python
df.dtypes
```

```python
df.astype({'col_1': 'float', 'col_2': 'float'})
```

```python
pd.to_numeric(df.col_3, errors='coerce')
```

```python
pd.to_numeric(df.col_3, errors='coerce').fillna(0)
```

```python
df = df.apply(pd.to_numeric, errors='coerce').fillna(0)
df
```

## Reduce DataFrame Size

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.info(memory_usage='deep')
```

```python
# Use smaller number of columns or read object data types as categorical
drinks = pd.read_csv('http://bit.ly/drinksbycountry', dtype={'continent': 'category'})
drinks.info(memory_usage='deep')
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry',
                     dtype={'continent': 'category'},
                     usecols=['beer_servings', 'continent'])
drinks.info(memory_usage='deep')
```

## Styling

```python
stocks.head()
```

```python
format_ = {
    'Date': '{:%m/%d/%y}', 'Close': '${:.2f}', 'Volume': '{:,}'
}
(stocks.style.format(format_)
 .hide_index()
 .highlight_min(subset='Close', color='red')
 .highlight_max(subset='Close', color='lightgreen')
 .background_gradient(subset='Volume', cmap='Greens')
 .set_caption('Stock Data Frame')
)
```

```python
format_ = {
    'Date': '{:%m/%d/%y}', 'Close': '${:.2f}', 'Volume': '{:,}'
}
(stocks.style.format(format_)
 .hide_index()
 .highlight_min(subset='Close', color='red')
 .highlight_max(subset='Close', color='lightgreen')
 .bar(subset='Volume', align='zero', color='lightblue')
 .set_caption('Stock Data Frame')
)
```
